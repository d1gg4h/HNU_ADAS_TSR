## Verarbeitung und Manipulation

Die Manipulation und Verarbeitung von Bildern bildet die Grundlage für unterschiedliche Aspekte im Maschinellen Sehen sowie Lernen.<br>

### Filter im Maschinellen Lernen

Im Umfeld des *machine learning* bieten Filter die Möglichkeit Bilder vielschichtig zu bearbeiten. Dieser Vorgang wird *pre-processing*, also Vorverarbeitung, genannt.
Einerseits können bestimmte Bereiche verstärkt dargestellt werden, hier insbesondere Kanten von Objekten. Andererseits können mit Filtern ungewünschte Störungen, wie Bildrauschen (*noise*) vermindert werden.
Dies dient dem Zweck, die Genauigkeit und die Erkennungsrate zu erhöhen.

Ein Beispiel für einen geläufigen Filter ist der *Gauß-Filter*.
Dieser lässt sich als 3x3 Matrix wie folgt darstellen:

Im Falle des Gauß-Filters wird das Bild mit einem *Weichzeichner* entschärft, dabei wird störendes Bildrauschen unterdrückt sowie der Moire-Effekt verringert.

Wenn nun ein Filter auf die Eingabe angewendet wird, also eine Faltungsoperation abläuft, verändert sich die Eingangsmatrix (bzw. das Eingangsbild).

````python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
%matplotlib inline

image = cv2.imread('Einstein.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) # Umwandlung der Farben von RGB zu HSV

blur1_image = cv2.medianBlur(image, 3)
blur2_image = cv2.medianBlur(image, 9)
plt.figure(figsize=(15,6))
plt.subplot(131), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(cv2.cvtColor(blur1_image, cv2.COLOR_HSV2RGB)),plt.title('Mean filter')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(cv2.cvtColor(blur2_image, cv2.COLOR_HSV2RGB)),plt.title('Mean filter')
plt.xticks([]), plt.yticks([])
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma)),plt.title('Normalverteilung')
plt.show()
````

````python
gauss1_image = cv2.GaussianBlur(image, (3, 3),0)
gauss2_image = cv2.GaussianBlur(image, (9, 9),0)
plt.figure(figsize=(15,6))
plt.subplot(131), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB)),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(cv2.cvtColor(gauss1_image, cv2.COLOR_HSV2RGB)),plt.title('3x3 Gauss-Filter')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(cv2.cvtColor(gauss2_image, cv2.COLOR_HSV2RGB)),plt.title('9x9 Gauss-Filter')
plt.xticks([]), plt.yticks([])
plt.show()
````

#### Laplace-Filter
Dieser Filter wird für die Kantendetektion verwendet. Kanten sind als starke Gradienten der Helligkeit im Bild zu erkennen, das heißt, den Verlauf der Werte mit der höchsten Änderungsrate. Da wo sich die Farben schnell und stark verändern, wird eine Kante festgestellt. Ein möglicher Laplace-Filter sieht wie folgt aus:

````python
image2 = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

laplace_image = cv2.Laplacian(image2,cv2.CV_64F)
plt.figure(figsize=(15,6))
plt.subplot(131), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(laplace_image, cmap='gray'),plt.title('Laplacian')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(image2 + laplace_image, cmap='gray'),plt.title('Resulting image')
plt.xticks([]), plt.yticks([])
plt.show()
````

Wir erkennen, dass die Kanten zwar verschärft wurden, das Bildrauschen jedoch auch stark verschärft wurde.
Können wir das Ergebnis vielleicht verbessern?

````python
gauss1_image = cv2.GaussianBlur(image2, (3, 3),0)
laplace3_image = cv2.Laplacian(gauss1_image,cv2.CV_64F)
print("Kombination von Gauß- und Laplace-Filter")
show(laplace3_image)
````

#### Sobel Filter
Dies ist ebenfalls ein Kantendetektions-Filter, der mithilfe der ersten Ableitung der Helligkeitswerte arbeitet. Dabei kann die Ableitung von x oder y erfolgen.

````python
sobelx = cv2.Sobel(image2,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(image2,cv2.CV_64F,0,1,ksize=5)

plt.figure(figsize=(17,6))
plt.subplot(141), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(142), plt.imshow(sobelx, cmap='gray'),plt.title('Sobel-X')
plt.xticks([]), plt.yticks([])
plt.subplot(143), plt.imshow(sobely, cmap='gray'),plt.title('Sobel-Y')
plt.xticks([]), plt.yticks([])
plt.subplot(144), plt.imshow(laplace3_image, cmap='gray'),plt.title('Gauss+Laplace-Filter')
plt.xticks([]), plt.yticks([])
plt.show()
````

### Datenaugmentation
Ein weiteres Einsatzgebiet der Manipulation und Verarbeitung der Bilder ist die Datenaugmentation mit Hilfe von Bildverändernden Mitteln.<br>
Bei der Datenaugmentation generiert man zusätzliche Bilder aus einer begrenzten Datenbasis durch deren Veränderung. 
Hintergrund dieser Aktion ist das Problem der Überanpassung. Dabei wird das *machine learning* Model zu sehr an die Trainingsdaten angepasst und lässt sich nicht mehr allgemein auf andere, neue, Daten anwenden.<br>
<p style="color:white;background-color:#85C1E9;">Beispiel Überanpassung:<br> Ihr lernt das Model mit 10000 Bildern eures Schäferhundes an, dabei wird das Model ab einem gewissen Grad nur noch euren Hund (anhand unikaler Merkmale) erkennen, jedoch keine anderen Schäferhunde.</p><br>
In einem optimalen Fall stünden einem Modell eine unerschöpfliche Datenmenge zur Verfügung, aus der unendlich unterschiedliche Muster erlernt werden können.
Da in der Realität nur eine begrenzte Datenmenge existiert und man einer Überanpassung entgegenwirken möchte, kann man seine Datenbasis mit Hilfe der Augmentation erweitern.

Folgende Möglichkeiten bieten sich an:

- Rotation
- Spiegelung
- Verzerrung
- Veränderung mit Filtern (Gauß, etc.)
- Zufälliges Bilderrauschen (*random noise*)

````python
from scipy import ndimage

flip_ud_face = np.flipud(img)

rotate_face = ndimage.rotate(img, 45)
rotate_face_noreshape = ndimage.rotate(img, 45, reshape=False)

plt.figure(figsize=(15,6))
plt.subplot(131), plt.imshow(flip_ud_face, cmap='gray'),plt.title('180°')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(rotate_face, cmap='gray'),plt.title('45° und angepasst')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(rotate_face_noreshape, cmap='gray'),plt.title('45° nicht angepasst')
plt.xticks([]), plt.yticks([])
plt.show()
````
